### Personal Income
- Personal income consists of the income that persons receive in return for their provision of labor, land, and capital used in current production as well as other income, such as personal current transfer receipts
- In the state and local personal income accounts the personal income of an area represents the income received by or on behalf of the persons residing in that area
- It is calculated as the sum of wages and salaries, supplements to wages and salaries, proprietors' income with inventory valuation (IVA) and capital consumption adjustments (CCAdj), rental income of persons with capital consumption adjustment (CCAdj), personal dividend income, personal interest income, and personal current transfer receipts, less contributions for government social insurance plus the adjustment for residence

```python
# making a function to speed up the process
def bea_transform(df, value_name, value_mapping=value_mapping):
    df = df.melt(
        id_vars=['GeoFIPS', 'GeoName'],   # columns to keep
        var_name='Year',                  # name for the new year column
        value_name=value_name   
    )
    df['GeoName'] = df['GeoName'].replace(value_mapping) # Changing state names to abbreviations
    df['LocYear'] = df['GeoName'].astype(str) + df['Year'].astype(str) # Combining state names and years into new index column called LocYear
    df = df.drop(columns=["GeoName", "Year"]) # dropping unneeded columns
    return df
```

## Fixing Data Types
