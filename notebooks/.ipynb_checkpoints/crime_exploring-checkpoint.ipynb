```python
# making a function to make everything easier

def cleaning_crime(filepath, space=' '):
    df = pd.read_excel(
        filepath,
        engine="xlrd",
        sheet_name=0,
        skiprows=3
    )
    # cleaning column names + dropping empty columns
    df.columns = df.columns.str.replace('\n', space, regex=False)
    df = df.dropna(axis=1, how='all')
    # forward filling the state and area data (it is missing otherwise)
    df['State'] = df['State'].ffill()
    df['Area'] = df['Area'].ffill()
    # only keeping the state total rows
    df = df[
        (df['Area'] == 'State Total') &
        (df['Unnamed: 2'] == "Rate per 100,000 inhabitants")
    ].copy()
    # removing footnote
    df = df[df['State'].str.len() < 20]
    # idk what these columns do...hopefully they are not important
    df = df.drop(columns=['Area', 'Unnamed: 2', 'Population'])
    # getting rid of footote numbers
    df['State'] = df['State'].str.replace(r',', '', regex=True).str.strip()
    df['State'] = df['State'].str.replace(r'\d+$', '', regex=True).str.strip()
    df['State'] = df['State'].replace(" ", "")
    df = df.reset_index(drop=True)
    
    return df
```

## Single LR

```python
def r2_scores(X, y):
    r2_dict = dict()
    for col_name in X:
        X2 = X[[col_name]]
        model = LinearRegression()
        model.fit(X2, y)
        r_sq = model.score(X2, y)
        r2_dict[col_name] = r_sq
        r_sq = None
    return r2_dict
```

```python
single_scores = r2_scores(X, y)
```

```python
single_scores
```

```python
single_lr = pd.DataFrame(list(single_scores.items()), columns=['Feature', 'SingleLinearReg'])
single_lr
```

```python
plt.figure(figsize=(13, 6))
plt.bar(single_scores.keys(), single_scores.values(), color="maroon")
plt.xlabel("Feature Names", fontsize=12)
plt.ylabel("r^2 score", fontsize=12)
plt.xticks(fontsize=10)
plt.title("New Crime Single-Variable Prediction Results for Linear Regression", fontsize=15)
plt.show()
```

## Permutation Feature Importance

```python
from sklearn.model_selection import train_test_split
from sklearn.inspection import permutation_importance
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42)

rf = RandomForestRegressor(random_state=42).fit(X_train, y_train)
print(f"Model Score: {rf.score(X_test, y_test)}")

r = permutation_importance(rf, X_test, y_test,
                           n_repeats=30,
                           random_state=0, scoring='r2')

column_names_list = X.columns.tolist()

for i in r.importances_mean.argsort()[::-1]:
    print(f"{column_names_list[i]:<8}" + ': '
            f"{r.importances_mean[i]:.3f}"
            f" +/- {r.importances_std[i]:.3f}")
    
feature_importance_df1 = pd.DataFrame({'Feature': column_names_list, 'Permutation': r.importances_mean})    
print("Feature importance:\n", feature_importance_df1)
feature_importance_df1
```

```python
plt.figure(figsize=(13, 6))
plt.bar(feature_importance_df1["Feature"], feature_importance_df1["Permutation"], color="firebrick")
plt.xlabel("Feature Names", fontsize=12)
plt.ylabel("Permutation Score", fontsize=12)
plt.xticks(fontsize=10)
plt.title("Crime Permutation Feature Importance Scores", fontsize=15)
plt.show()
```

```python
def get_rankings_og(df_list, column_list):
    D = {
        "Features" : ["PI_rate", "% >= Bachelor's", "House Leaning", "Robbery", "Aggravated assault", "Burglary", "Property crime"], 
        "Scores": [0., 0., 0., 0., 0., 0., 0.]
    }
    ranking_df = pd.DataFrame(D)
    ranking_df = ranking_df.set_index('Features')
    col_idx = 0 
    for df in df_list:
        index = 0 
        place = 4
        for i in range(1, len(df[column_list[col_idx]])):
            prev = df[column_list[col_idx]][index]
            curr = df[column_list[col_idx]][i]
            diff = prev - curr
            place_scale = place + diff
            ranking_df.loc[df["Feature"][index]] += place_scale
            index +=1
            place -= 1
        ranking_df.loc[df["Feature"][index]] += 1
        col_idx += 1
    return ranking_df
```

```python
rankings_og = get_rankings_og(df_list, column_list).sort_values(by="Scores", ascending=False)
```
