## Data Gathering + Cleaning: Crime
- For the following section, I will be getting all the data from [Crime/Law Enforcement Stats (Uniform Crime Reporting Program)](https://www.fbi.gov/how-we-can-help-you/more-fbi-services-and-information/ucr)

```python
import pandas as pd
```

```python
df = pd.read_excel(
    "data/crime/2010.xls",
    engine="xlrd",
    sheet_name=0,
    skiprows=3
)

# Clean column names
df.columns = df.columns.str.replace('\n', ' ', regex=False)

# Drop empty columns
df = df.dropna(axis=1, how='all')

# Forward-fill State and Area
df['State'] = df['State'].ffill()
df['Area'] = df['Area'].ffill()

# Keep only State Total rows (NaN or blank in Unnamed: 2)
df = df[
    (df['Area'] == 'State Total') &
    (df['Unnamed: 2'].isna() | (df['Unnamed: 2'].str.strip() == ''))
].copy()

# Remove footnote rows
df = df[df['State'].str.len() < 20]

# Clean up columns
df = df.drop(columns=['Area', 'Unnamed: 2', 'Unnamed: 13'])

# Clean footnote numbers from state names (ILLINOIS2 → ILLINOIS)
df['State'] = df['State'].str.replace(r'\d+$', '', regex=True).str.strip()

df = df.reset_index(drop=True)
```
